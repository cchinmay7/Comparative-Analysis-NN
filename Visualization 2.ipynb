```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['MNIST', 'Extended MNIST', 'Fashion MNIST', 'CIFAR-10']
models = [
    '5 epochs - 512 nodes', '5 epochs - 1024 nodes', '50 epochs - 512 nodes', '50 epochs - 1024 nodes',
    '5 epochs - 512 nodes', '5 epochs - 1024 nodes', '50 epochs - 512 nodes', '50 epochs - 1024 nodes'
]
colors = ['red', 'blue', 'red', 'blue', 'yellow', 'green', 'yellow', 'green']
hatches = ['', '', '////', '////', '', '', '////', '////']  

data = np.array([
    [0.9784, 0.9837, 0.9809, 0.9845],
    [0.8372, 0.8210, 0.8428, 0.8356],
    [0.8756, 0.8939, 0.8711, 0.8944],
    [0.4483, 0.4860, 0.4548, 0.4988]
])


additional_data = np.array([
    [0.9886, 0.9930, 0.9924, 0.9915],
    [0.8714, 0.8573, 0.8648, 0.8511],
    [0.8848, 0.8999, 0.8928, 0.9113],
    [0.6392, 0.6814, 0.6564, 0.6901]
]) 


combined_data = np.concatenate((data, additional_data), axis=1)


fig, ax = plt.subplots(figsize=(12, 6))


bar_width = 0.1
x_positions = np.arange(len(categories))

for i, model in enumerate(models):
    if i >= 4:  
        bar_position = x_positions + (i * bar_width) + bar_width / 2
    else:
        bar_position = x_positions + (i * bar_width)
    ax.bar(bar_position, combined_data[:, i], bar_width, color=colors[i], hatch=hatches[i], label=model)


ax.set_xlabel('Categories')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy Comparison of various models with Different Datasets')
ax.set_xticks(x_positions + (bar_width * 4))
ax.set_xticklabels(categories)
ax.set_yticks(np.arange(0.0, 1.05, 0.1))

handles, labels = ax.get_legend_handles_labels()
new_labels = [
    'DNN - 5 epochs - 512 nodes', 'DNN - 50 epochs - 512 nodes', 'DNN - 5 epochs - 1024 nodes', 'DNN - 50 epochs - 1024 nodes',
    'CNN - 5 epochs - 512 nodes', 'CNN - 50 epochs - 512 nodes', 'CNN - 5 epochs - 1024 nodes', 'CNN - 50 epochs - 1024 nodes'
]
ax.legend(handles, new_labels, loc='upper left', bbox_to_anchor=(1, 1))


plt.tight_layout()
plt.show()
```
