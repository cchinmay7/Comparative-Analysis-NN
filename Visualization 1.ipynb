```python
import numpy as np
import matplotlib.pyplot as plt

epochs = [5, 50]

accuracy_thin_red = [0.9784, 0.9837]
accuracy_thin_blue = [0.9809, 0.9845] 
accuracy_thick_red = [0.9886, 0.9930] 
accuracy_thick_blue = [0.9924, 0.9915] 


plt.figure(figsize=(10, 6))


plt.plot(epochs, accuracy_thick_red, color='red', linewidth=3, label='CNN-512 nodes')
plt.plot(epochs, accuracy_thick_blue, color='blue', linewidth=3, label='CNN-1024 nodes')
plt.plot(epochs, accuracy_thin_red, color='red', linewidth=2, linestyle='dashed', label='DNN-512 nodes')
plt.plot(epochs, accuracy_thin_blue, color='blue', linewidth=2, linestyle='dashed', label='DNN-1024 nodes')


min_val = min(min(accuracy_thick_red), min(accuracy_thick_blue), min(accuracy_thin_red), min(accuracy_thin_blue))
max_val = max(max(accuracy_thick_red), max(accuracy_thick_blue), max(accuracy_thin_red), max(accuracy_thin_blue))


plt.scatter(5, min_val, color='red', s=100, marker='x', label='Lowest Accuracy')
plt.scatter(50, max_val, color='red', s=100, marker='*', label='Highest Accuracy')




plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.xticks(epochs)
plt.yticks([0.970, 0.975, 0.98, 0.985, 0.99, 0.995, 1.00])

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.suptitle('Accuracy Comparison for MNIST',x=0.45, y=0, fontsize=16)

plt.tight_layout()
plt.show()
```
